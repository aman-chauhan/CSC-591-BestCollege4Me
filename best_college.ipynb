## Filling NaN values

```python
df['RELAFFIL'] = df['RELAFFIL'].fillna(-1)

def fill_nan(col):
    if df[col].dtype != 'O':
        df[col].fillna(df[col].median(), inplace=True)

for col in cols:
    fill_nan(col)
```
