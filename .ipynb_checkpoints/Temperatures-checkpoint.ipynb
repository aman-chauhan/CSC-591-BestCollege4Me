```python
fips=json.load(open('zip2fips.json'))
```

```python
def getTemps(zipcode):
    data=[]
    rp=dict(params)
    if zipcode in fips:
        rp['locationid']='FIPS:'+fips[zipcode]
    else:
        rp['locationid']='ZIP:'+zipcode
    for season in sorted(seasons.keys()):
        rp['startdate']=seasons[season]['start']
        rp['enddate']=seasons[season]['end']
        rp['offset']=0
        r=requests.get(url,headers=headers,params=rp)
        d=r.json()
        if len(d.keys())==0:
            data.append(None)
            continue
        if 'status' in d.keys():
            data.append(None)
            continue
        if d['metadata']['resultset']['count']>d['metadata']['resultset']['limit']:
            if int(d['metadata']['resultset']['count']) <= 1000:
                rp['limit']=int(d['metadata']['resultset']['count'])
            else:
                rp['limit']=1000
            time.sleep(0.5)
            r=requests.get(url,headers=headers,params=rp)
            d=r.json()
        avg=0.0
        cnt=0
        flag=0
        if 'results' not in d.keys():
            data.append(None)
            continue
        for station in d['results']:
            if station['datatype']=='TAVG':
                flag=1
                avg+=float(station['value'])
                cnt+=1
        time.sleep(0.5)
        if flag==0:
            data.append(None)
        else:
            data.append(((avg/cnt)*9/5)+32)
    return data
```

```python
df=pd.read_csv('cleaned_data 2.csv')
```

```python
zips=df['ZIP'].tolist()
zips=[zips[i].split("-")[0] for i in range(len(zips))]
```

```python
def noneCount(ele):
    cnt=0
    for x in ele:
        if x is None:
            cnt+=1
    return cnt
```

```python
with open ('temps_try1.pickle', 'rb') as fp:
    temps = pickle.load(fp)
```

```python
for i in range(len(zips)):
    if noneCount(temps[i])!=0:
        t=getTemps(zips[i])
        if noneCount(t)<noneCount(temps[i]):
            temps[i]=t
            print(str(i)+' - '+zips[i]+' improved.')
        else:
            print(str(i)+' - '+zips[i]+' not improved.')
```

```python
with open('temps_try1.pickle', 'wb') as fp:
    pickle.dump(temps, fp)
```

```python
cols=np.asarray(temps,dtype=float)
```

```python
states=df['STABBR'].tolist()
st_avg={x:{'value':[0,0,0,0], 'count':0} for x in df['STABBR'].unique().tolist()}
for i in range(len(zips)):
    if noneCount(temps[i])==0:
        st_avg[states[i]]['count']+=1
        for j in range(4):
            st_avg[states[i]]['value'][j]+=temps[i][j]
for x in st_avg:
    st_avg[x]['value']=[y/st_avg[x]['count'] if st_avg[x]['count']!=0 else 0 for y in st_avg[x]['value']]
for i in range(len(zips)):
    for j in range(4):
        if temps[i][j] is None:
            temps[i][j]=st_avg[states[i]]['value'][j]
v={zips[i]:temps[i] for i in range(len(zips))}
with open('temperatures.json','w') as fp:
    json.dump(v,fp)
```
